#### Calculating the percentage of missing values

```python
n1 = df['Type2'].isnull().sum()
n2 = df.shape[0]
per_val = abs((n1/n2) * 100)
print(per_val)
```

### Finding the missingness mechanism
### 1.MCAR (Missing Completely At Random): Missingness unrelated to any data.
### 2.MAR (Missing At Random): Missingness depends on observed data but not unobserved.
### 3.MNAR (Missing Not At Random): Missingness depends on the missing value itself (hardest).
### As of now it could be a possibitity of MCAR and The "Type2" column is an categorical columns whereas Mean and median wouldn't be possible. Mode is more suitable But it could Artificially reduces the standard deviation
### The best way of handling the missing values here using the Forward fill and Backward fill

```python
df["Type2"].ffill(inplace=True)
```
