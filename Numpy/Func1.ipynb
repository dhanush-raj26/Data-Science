Loop

```python
import numpy as np 
arr = np.array([1,2,3,4,5,6])

for i in np.nditer(arr):
    print(i,end=" ")
```

```python
for i,j in np.ndenumerate(arr):
    print(i,j,end=" ")
```

views vs copies

```python
view = arr[:2]
view[0] = 100
print(arr)
```

```python
copies = arr.copy()
copies[0] = 200
print(arr)
```
