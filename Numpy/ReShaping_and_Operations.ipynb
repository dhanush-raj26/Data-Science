Reshaping

```python
import numpy as np
arr = np.array([1,2,3,4,5,6])
arr.reshape(2,3)
arr.reshape(6,1)
```

Ravel

```python
arr = np.array([1,2,3,4,5,6])
arr.reshape(2,3)
arr.reshape(6,1)
arr.ravel() ## convert the multi dimensional array into single dimensional array
```

Flatten

```python
### flatten is also used to convert multi dimensional array into single dimensional array 
### but it will return the copy of original array where as ravel will return the view of original array.
arr = np.array([1,2,3,4,5,6])
t = arr.reshape(2,3)
```

```python
t.flatten()
# t
```

Arithmetic operations

```python
arr1 = np.array([1,2,3,4,5,6])
arr2 = np.array([7,8,9,10,11,12])
print(arr1 + arr2) ### element wise addition
print(arr1 - arr2) ### element wise subtraction
print(arr1 * arr2) ### element wise multiplication
print(arr1 / arr2) ### element wise division
print(arr1 // arr2) ### element wise floor division
print(arr1 % arr2) ### element wise modulus
```

Universal Function ---> ufuncs

```python
arr = np.array([1,4,9,16])
np.sqrt(arr) 
```
