```python
import numpy as np

arr = np.array(["dhanus",1,2,3,5,6])
arr.dtype
```

```python
arr = np.array([1,2,3,5,6]) ### bydefault it will take int64
arr.dtype
```

```python
arr = np.array([1,2,3,5,6],dtype=np.float64)
arr.dtype
```

Type Casting

```python
arr = np.array([1,2,3,5,6])
new_arr = arr.astype(np.float64)
new_arr
```

```python
arr = np.array([True,False,True])
new_arr = arr.astype(np.int64)
new_arr
```
