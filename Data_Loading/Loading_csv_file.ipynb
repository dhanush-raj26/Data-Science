1. Read specific columns using read_csv

```python
temp = pd.read_csv("real_drug_dataset.csv",usecols=["Age","Gender"])
temp.head()
```

2.Setting an Index Column (index_col)

```python
temp = pd.read_csv("real_drug_dataset.csv",index_col=0)
temp.head()
```

3. Handling Missing Values Using read_csv

```python
temp = pd.read_csv("real_drug_dataset.csv",na_values=["N/A", "Unknown"])
temp.head()
```

4. Using skiprows in read_csv()

```python
temp = pd.read_csv("real_drug_dataset.csv",skiprows=[2,3,4,5,6])
temp.head()
```
